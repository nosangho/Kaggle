```python
import pandas as pd
reviews = pd.read_csv('./winemag-data_first150k.csv', index_col=0)
reviews.head()
```

```python
import matplotlib.pyplot as plt
fig, axarr = plt.subplots(2, 1, figsize=(12,8))
```

```python
axarr
```

```python
fig, axarr = plt.subplots(2, 1, figsize=(12,8))
reviews['province'].value_counts().head(20).plot.bar(ax=axarr[0], color='royalblue')
reviews['points'].value_counts().sort_index().plot.bar(ax=axarr[1], color='firebrick')
```

```python
fig, axarr = plt.subplots(2,2, figsize=(12,8))
```

```python
fig, axarr = plt.subplots(2,2, figsize=(12,8))
reviews['points'].value_counts().sort_index().plot.bar(ax = axarr[0][0], color='royalblue')
reviews['province'].value_counts().head(20).plot.bar(ax = axarr[1][1], color='royalblue')
```

```python
fig, axarr = plt.subplots(2,2, figsize=(12,8))

title_font_size = 16

reviews['points'].value_counts().sort_index().plot.bar(ax=axarr[0][0], color='lightblue')
axarr[0][0].set_title('Wine Score', fontsize=title_font_size)

reviews['variety'].value_counts().head(20).plot.bar(ax=axarr[1][0], color='royalblue')
axarr[1][0].set_title('Wine Varieties', fontsize=title_font_size)

reviews['price'].value_counts().head(20).plot.bar(ax=axarr[0][1], color='indianred')
axarr[0][1].set_title('Wine Price', fontsize=title_font_size)

reviews['province'].value_counts().head(20).plot.bar(ax=axarr[1][1], color='coral')
axarr[1][1].set_title('Wine Origin', fontsize=title_font_size)

plt.subplots_adjust(hspace=.5)

import seaborn as sns
sns.despine()
```
